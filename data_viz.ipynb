#### Load the data you receive into a Pandas DataFrame

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import seaborn as sns
```

```python
startups = pd.read_csv("50_Startups.csv")
```

#### Show the first five rows of the data set

```python
startups.head(5)
```

#### Show the description and the info of the data set

```python
startups.describe()
```

```python
startups.info()
```

#### Visualize your dataset to demonstrate insights into the data

```python
sns.set()
plt.figure(figsize=(10,6),dpi=300)
sns.regplot(x='R&D Spend', y='Profit', data=startups, order=1)
```

#### Verbalize your insights at bottom of notebook

What I first see when looking at this visualization, there seems to be a strong correlation between the amount of money that you spend in R&D in a startup, and the amount of profit you will make. But when I start to take a closer look, I realize that the companies that spent next to nothing on R&D still made a profit.  And then I thought about the fact that this only says "Profit" not "Profit Margin" so on a closer look, the companies that spent very little on R&D might have actually made a larger profit margin.  This also doesn't take into account other expenses, like employment costs and advertising.  So while at first glance, you can see a correlation, there is no reason to believe that spending more on R&D will lead you to a larger profit.

```python
sns.set(style="darkgrid")
graph = sns.jointplot("points", "price", data=wine, kind="reg",
                  xlim=(80, 100), ylim=(0, 500), color="b", height=10)
plt.show(graph)
```

```python
sns.set()
plt.figure(figsize=(10,6),dpi=300)
sns.regplot(x='points', y='price', data=wine, order=1)
```
